## Teste de Hipótese

### H<sup>null</sup>
- A distribuição da taxa da gorjeta é a mesma nos dois grupos (Pediram ou não sobremesa)

### H<sup>alt</sup>
- A distribuição da taxa da gorjeta não é a mesma nos dois grupos (Pediram ou não sobremesa)

```python
from scipy.stats import ranksums
```

```python
dessert = df.loc[df["dessert"] == 'Yes']["percent"]
not_dessert = df.loc[df["dessert"] == 'No']["percent"]
```

```python
r = ranksums(dessert, not_dessert)
print(f"PVALUE: {r.pvalue}")
```

pvalue < 0.05 aceitamos a hipótese alternativa, nesse caso vamos com a hipótese nula

- A distribuição da taxa da gorjeta é a mesma nos dois grupos (Pediram ou não sobremesa)

## Analisando a média das gorjetas por dia


```python
mean = df["tip"].mean()
print(f"A média das gorjetas é de: {mean}")
```

```python
df.groupby("day").mean()
```

```python
df["day"].value_counts()
```

## teste de Hipótese

**H<sup>null</sup>**: A distribuição da conta é igual no sábado e no domingo

**H<sup>alt</sup>**: A distribuição da conta não é igual no sábado e no domingo

Aceita a hipótese alternativa se `p_value < 0.05`


```python
valor_sabado = df.loc[df["day"] == "Sat"]["total_bill"]
valor_domingo = df.loc[df["day"] == "Sun"]["total_bill"]
```

```python
r2 = ranksums(valor_domingo, valor_sabado)
print(f"O valor do pvalue é: {r2.pvalue}")
```

**H<sup>alt</sup>**: A distribuição da conta não é igual no sábado e no domingo
